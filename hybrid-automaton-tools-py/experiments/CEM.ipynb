```python
# ipython nbconvert CEM.ipynb --to=latex --post=pdf --template=latex_nocode.tplx --SphinxTransformer.author='Clemens Eppner'
%matplotlib inline

import matplotlib.pyplot as plt
import yaml
import numpy as np
import glob

def load_yamls(filenames):
    files = sorted(glob.glob(filenames))
    res = []
    for f in files:
        res.append(yaml.load(open(f, 'r')))
    return res
```

Let's look at the success rate over time, after iteratively applying the Cross-Entropy Method.


```python
robot_yamls = [load_yamls(s + "/*yml") for s in ["cem_apple/iter{:02d}".format(i+1) for i in range(10)]]
robot_runs = np.vstack([[int(y['label']) for y in ys] for ys in robot_yamls])

parameters = {}
parameters["wallforce [N]"] = np.vstack([[(y['wall_force']) for y in ys] for ys in robot_yamls])
parameters["speed [m/s]"] = np.vstack([[(y['sliding_speed']) for y in ys] for ys in robot_yamls])
parameters["inflation [kPa]"] = np.vstack([[(y['finger_inflation']) for y in ys] for ys in robot_yamls])
parameters["angle [rad]"] = np.vstack([[(y['angle_of_attack']) for y in ys] for ys in robot_yamls])
parameters["positionx [m]"] = np.vstack([[(y['position_offset_x']) for y in ys] for ys in robot_yamls])

plt.figure(figsize=(10,5))
plt.bar(np.arange(10)+0.6, np.mean(robot_runs, axis=1))
plt.ylabel("Probability of success")
plt.xlabel("Iteration")
plt.xlim(0.5, 10.5)
plt.ylim(0, 1.)
plt.title("Success Rate of Cross-Entropy Method per Iteration");
```

Next, we look at how the parameters evolved over time.

```python
for k, v in parameters.iteritems():
    plt.figure(figsize=(10,5))
    plt.title("Evolution of Angle of Attack of Wall Grasp (Real World)")
    plt.fill_between(np.arange(v.shape[0]), np.mean(v, axis=1) + 2.*np.std(v,axis=1), np.mean(v,axis=1) - 2.*np.std(v,axis=1), facecolor='b', alpha=0.35)
    plt.plot(np.sort(v), 'r')
    plt.plot(np.mean(v, axis=1), 'b', linewidth=3)
    print "Mean: ", np.degrees(np.mean(v, axis=1)[-1])
    plt.ylim([0, 3.0])
    plt.ylabel(k)
    plt.xlabel("Iteration")
```

We compare these results with the ones from simulation. Here, we only varied a single parameter: the angle of attack of the wrist. We evaluated each parameter setting in 16 scenarios: 2 objects (cylinder, sphere) * 2 positions_x * 2 positions_y * 2 orientations. In each iteration we sampled ten parameter values, which results in 160 for each iteration.

Let's look at the evolution of the wrist angle parameter.

```python

params_sim = {}
params_sim["angle [rad]"] = sim_runs[:,3:]
samples = np.loadtxt(basedir + "cem_angle_wall_runs_actions.txt")

for k, v in params_sim.iteritems():
    plt.figure(figsize=(10,5))
    plt.title("Evolution of Angle of Attack of Wall Grasp")
    mu = np.mean(samples, axis=1) # = v[:,0]
    std = np.std(samples, axis=1) # = v[:,1]
    plt.fill_between(np.arange(v.shape[0]), mu + 2.*std, mu - 2.*std, facecolor='b', alpha=0.35)
    print "Mean: ", np.degrees(mu[-1])
    plt.plot(samples, 'r')
    plt.plot(mu, 'b', linewidth=3)
    plt.ylabel(k)
    plt.xlabel("Iteration")
```
